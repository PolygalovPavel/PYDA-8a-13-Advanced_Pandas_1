### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head(100)
```

```python
log['source_type'] = log.apply(lambda x: x['traffic_source'], axis=1)
log.loc[(log.region.str.lower() == 'russia')&((log.traffic_source.str.lower() == 'paid')|(log.traffic_source.str.lower() == 'email')), 'source_type'] = 'ad'
log.loc[(log.region.str.lower() != 'russia')&((log.traffic_source.str.lower() == 'paid')|(log.traffic_source.str.lower() == 'email')), 'source_type'] = 'other'
log.loc[(log.traffic_source.str.lower() == 'yandex')|(log.traffic_source.str.lower() == 'google'), 'source_type'] = 'organic'
log.head()
```

### Домашнее задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. 

Леммы должны иметь строковый тип.

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
data.head()
```

```python
pip install pymystem3
```

```python
from pymystem3 import Mystem
```

```python
m = Mystem()
data['lemmas'] = data.apply(lambda x: "".join(m.lemmatize(x['keyword'])).strip(), axis=1)
data.head()
```
